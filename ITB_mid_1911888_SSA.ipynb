# 시각화3: 체중과 LDL 콜레스테롤의 관계 

```python
# 체중, 신장, LDL콜레스테롤만 추출, 비만도 계산 후 변수 추가.

df_L = df[['체중','신장', 'LDL콜레스테롤']]

df_L['비만도'] = round(df_L['체중'] / (df_L['신장'] / 100) **2, 1)

display(df_L)
```

```python
#LDL콜레스테롤 결측치 제거 

df_L = df_L.dropna(subset = ['LDL콜레스테롤'], axis=0)  

display(df_L)
df_L.info()
```

```python
df_L.plot(kind='scatter', x='비만도', y='LDL콜레스테롤')

plt.axhline(y=170, color='red', linewidth=1, alpha = 0.5)
plt.axvline(x=25, color='red', linewidth=1, alpha = 0.5)


# BMI가 18.5 이하면 저체중 
# 18.5 ~ 22.9 사이면 정상 
# 23.0 ~ 24.9 사이면 과체중 ／ 25.0 이상부터는 비만
```

**- 인사이트** : 일반적인 상식에 따르면 콜레스테롤 농도와 체중, 비만도는 비례한다. 

하지만 이 조사결과에서는 오히려 비만이 아닌 사람들이 높은 LDL콜레스테롤을 기록하고 있다는 점을 파악할 수 있다. 

비만도(BMI)가 25 이상일 경우 비만, LDL 콜레스테롤이 170mg/dl 보다 높을 경우 고LDL혈증으로 본다. 해당 산점도에서는 고LDL혈증인 사람들이 비만이 아닌 쪽에 몰려있다는 것을 알 수 있다.
